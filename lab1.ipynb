The first step is to read the data from a csv file. We do this by using the loadtxt function of the numpy library and specifying comma as a delimiter for our data. We also want to skip the first row, because it contains a header for our data.

```python
import numpy as np
import matplotlib.pyplot as plt

# loading data from csv files without the first row
data_train = np.loadtxt('lab_1_train.csv', delimiter=',', skiprows=1)
data_test = np.loadtxt('lab_1_test.csv', delimiter=',', skiprows=1)
```

The next step would be to assign the data that we read from a csv file into our matrix variables. The x values are stored in a second column, while y values are stored in the third column (first column stores ID's, that we don't need). We also want to reshape our 1D X array into a vector with 1 column and as many rows as we have in the csv file.

```python
#Task A
# loading x (second column) and y (third column)
x = data_train[:, 1].reshape(-1, 1) # reshape x to be a vector
y = data_train[:, 2] 

plt.scatter(x, y)

plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
x = np.hstack((np.ones_like(x), x)) # Adding a column of ones for the bias term
print(x)
```

The final preparation step is to set the learning rate (alpha), initilize weights to random numbers, set tolerance and the initial value for the cost function.

```python
alpha = 0.01 # setting learning rate
weights = np.random.rand(2) # initializing weights to random numbers
print(weights)
```

```python
epsilon = 0.000001
prev_cost = np.inf
iteration_count = 0
```

Now we need to train our data. To do this at each step we first need to find the partial derivate of the cost function with respect to the weights. We use the mean squared error (MSE) as the cost function. After finding the gradient of the cost function we use gradient descent algorithm to train our data. The idea is to take small steps in the direction that is opposite to gradient, meaning the direction in which function is decreasing. Then we calculate our cost function as see, whether the difference in cost functions between iterations is less than our specified tolerance. If so, we break from the cycle. Otherwise, we do the above mentioned steps as long as the condition is not met.

```python
#Task B
while True:
    iteration_count += 1
    gradient = x.T @ (x @ weights - y) / len(y) # calculate the value of the partial derivative
    weights -= alpha * gradient # update weights by making a step in the opposite direction of the gradient
    cost = np.sum((x @ weights - y)**2) / (2 * len(y)) # calculate cost function
    
    if (np.abs(cost - prev_cost) < epsilon): # exit condition
        break
        
    prev_cost = cost
    
    print('Iteration: ', iteration_count)
    print('Weights: ', weights)
    print('Cost function: ', cost)
```

The next step is to plot train and test data and the predicted values. The train data is represented in blue dots and the test data is represented in red dots. The regression line is represented as a green solid line.

```python
#Task d
#Loading train and test datasets from csv files
x = data_train[:, 1].reshape(-1, 1)
y = data_train[:, 2]
x_test = data_test[:, 1].reshape(-1, 1)
y_test = data_test[:, 2]

plt.scatter(x, y, color='blue') # train data
plt.scatter(x_test, y_test, color='red') # test data
x_line = np.linspace(0, 1, len(x)) #creating an evenly spaced numbers between 0 and 1
y_line = weights[0] + weights[1] * x_line # calculating y value using already calculated weights
plt.plot(x_line, y_line, color='green')
x_test = np.hstack((np.ones_like(x_test), x_test))
```

The final step is to calculate the value of the cost function across all iterations. We do this by using the formula for the mean squared error. 

```python
#Task c

x_test = data_test[:, 1].reshape(-1, 1)
x_test = np.hstack((np.ones_like(x_test), x_test))
y_prediction = x_test @ weights

cost_test = np.sum((y_prediction - y_test)**2) / (2 * len(y))
print(cost_test)
```
